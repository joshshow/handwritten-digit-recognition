```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
print(X_train[7])
```

```python
plt.imshow(X_train[7])
plt.title(f"Label is: {y_train[7]}")
plt.show()
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)/255
X_test = X_test.reshape(-1, 28, 28, 1)/255
```

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.callbacks import EarlyStopping
```

```python
tf.random.set_seed(2)

es = EarlyStopping(
    monitor = 'val_loss',
    patience = 3,
    restore_best_weights=True
)

model = Sequential([
    Input(shape=(28,28,1)),
    Conv2D(32, (3,3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3)),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(50, activation='relu'),
    Dense(50, activation='relu'),
    Dense(10, activation='linear')
])

loss_fn = SparseCategoricalCrossentropy(from_logits=True)

model.compile(loss=loss_fn, optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.2, callbacks=es)
```

```python
dloss, daccuracy = model.evaluate(X_test, y_test)
print(f"The Loss recorded is: {dloss}\n The accuracy recorded is {daccuracy}")
```

```python
model.save("my_mnist_cnn.keras")
```

```python
nonny = model(X_test)
y_pred = tf.nn.softmax(nonny).numpy()

plt.imshow(X_test[7])
plt.title(f"Actual: {y_test[7]}\n Predicted: {np.argmax(y_pred[7])}")
plt.show()

print(y_pred[7])
```
